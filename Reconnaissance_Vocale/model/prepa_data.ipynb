```python
import os
import json
import librosa
```

```python
PATH = "../data"
JSON_PATH = "data.json"
#correspond à 1 seconde par defaut dans librosa
SAMPLES_TO_CONSIDER = 22050 
```

```python

def extraire_info_audio(file_path,num_mfcc=13, n_fft=2048, hop_length=512):
    # Découper L'audio pour assurer la cohérence de la longueur entre les différents fichiers
    signal, sample_rate = librosa.load(file_path)

    # Le nombre d'echantillon doit être supérieur à un certain seuil 
    if len(signal) >= SAMPLES_TO_CONSIDER:
        # Assurer la cohérence de la longueur du signal
        signal = signal[:SAMPLES_TO_CONSIDER]

        # Extraire les MFCCs
        MFCCs = librosa.feature.mfcc(signal, n_mfcc = num_mfcc, n_fft = n_fft, hop_length = hop_length)
        return MFCCs
    return []
```

```python
#Extrait les MFCC du jeu de données musical et les enregistre dans un fichier json afin de d'entrainer et 
#de valider le modèle 
# num_mfcc est le nombre de coefficients à extraire
# n_ftt est l'intervalle pour appliquer la FFT. Mesuré en nombre d'échantillons
# hop_length est la fenêtre coulissante pour FFT. Mesuré en nombre d'échantillons


def prep_data(dataset_path, json_path) :

    # dictionnaire pour stocker le label du l'audion et les informations importantes et nom du fichier
    data = {
        "mapping": [],
        "labels": [],
        "MFCCs": [],
    }

    # Pour parcourir tous les dossiers et étiquétés chaque audio par son label
    for i, (dirpath, dirnames, filenames) in enumerate(os.walk(dataset_path)):

        if dirpath is not dataset_path:

            # sauvegarder le label qui est le bom du dossier
            label = dirpath.split("/")[-1]
            data["mapping"].append(label)
            print("dossier " + label)

            #Parcourir tous les fichiers du dossier pour extraire les informations importantes
            for f in filenames:
                file_path = os.path.join(dirpath, f)
                MFCCs = extraire_info_audio(file_path)
                if len(MFCCs) > 0 :
                    data["MFCCs"].append(MFCCs.T.tolist())
                    data["labels"].append(i - 1)
                    print("{}: {}".format(file_path, i - 1))

    with open(json_path, "w") as fp:
        json.dump(data, fp, indent=4)
```

```python

if __name__ == "__main__":
    prep_data(PATH, JSON_PATH)
```
